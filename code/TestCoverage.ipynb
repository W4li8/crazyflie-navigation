## Online coverage path planning using $\epsilon*$, obstacle-less case

```python
from global_nav import GlobalNav
from global_display import GlobalDisplay as disp
from global_simulator import Simulator
import random

# defining a grid an an area to cover
glonav = GlobalNav(300,200,30,20,5,5,False)
sim = Simulator(glonav,1,1,1)
glonav.cover_area(0,0,10,19)

# defines a simulation step
def step():
    sim.sense()
    sim.check_collisions()
    [wx,wy] = glonav.cover()
    nx, ny = sim.waypoint_line(wx,wy)
    glonav.update(nx,ny)

for i in range(250):
    if i%50 == 0:
        disp.showGrid(glonav,confidenceAsAlpha = True)
    step()

disp.showGrid(glonav,confidenceAsAlpha = True)
```
