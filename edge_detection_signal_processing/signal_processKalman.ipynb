```python
def process(window):
    t = 0
    flt = [window[0]]
    diff = [0]
    flt_2 = [0]
    thresh_plot = [0]

    for i in range(0,len(window)):
        measure = window[i]
        t = t + 1
        flt.append(alpha*measure+(1-alpha)*flt[t-1])
        diff.append(abs(flt[t]-flt[t-1]))
        flt_2.append(alpha_2*diff[i]+(1-alpha_2)*flt_2[t-1])
        if alpha_2*diff[i]+(1-alpha_2)*flt_2[t-2] > thresh:
            out = 0.2
        else:
            out = 0
        thresh_plot.append(out)

    #plt.plot(flt_2)
    plt.plot(window)
    plt.plot(thresh_plot)
```

```python
def process_ranger(window):
    alpha = 0.1
    thresh = 0.8
    t = 0
    flt = [window[0]]
    diff = [0]
    flt_2 = [0]
    thresh_plot = [0]
    
    thresh_val_plot = [thresh]

    for i in range(0,len(window)):
        measure = window[i]
        t = t + 1
        flt.append(alpha*measure+(1-alpha)*flt[t-1])
        if flt[-1] < thresh:
            out = 2
        else:
            out = 0
        thresh_plot.append(out)
        thresh_val_plot.append(thresh)

    plt.plot(flt)
    plt.plot(thresh_plot)
    plt.plot(thresh_val_plot)
    plt.plot(window)
```

```python
process_ranger(ranger_front[500:1200])
```
