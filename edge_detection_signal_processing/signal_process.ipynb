```python
alt = np.load("altitude_values.npy", allow_pickle=True)
plt.plot(alt)
```

```python
alpha = 0.001
t = 0
alpha_2 = 0.3
thresh = 5e-5
thresh_plot = [0]
```

```python
window = alt

#diff = np.diff(window,2)
#diff = np.abs(diff)


flt = [0.3]
diff = [0]
flt_2 = [0]

for i in range(0,len(window)):
    measure = window[i]
    t = t + 1
    flt.append(alpha*measure+(1-alpha)*flt[t-1])
    diff.append(abs(flt[t]-flt[t-1]))
    flt_2.append(alpha_2*diff[i]+(1-alpha_2)*flt_2[t-1])
    if alpha_2*diff[i]+(1-alpha_2)*flt_2[t-1] > thresh:
        out = 2e-5
    else:
        out = 0
    thresh_plot.append(out)


plt.plot(flt_2)
plt.plot(thresh_plot)
```
