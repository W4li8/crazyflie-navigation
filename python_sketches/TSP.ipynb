```python
import matplotlib.pyplot as plt
import networkx as nx
import dwave_networkx as dnx
import random
import dimod
```

# Trying out TSP approximation in Networkx

```python
# create a grid graph (as a map)
G = nx.Graph()
nodes = []
pos = {}
for x in range(10):
    for y in range(10):
        node = (x,y)
        G.add_node(node)
        nodes.append(nodes)
        pos[node] = [x,y]
        if x>0 :
            G.add_edge((x,y),(x-1,y))
        if y>0 :
            G.add_edge((x,y),(x,y-1))
            
# remove a bunch to approximate obstacles 
for i in range(10):
    G.remove_node((random.randrange(10),random.randrange(10)))

nx.draw(G, with_labels=True, pos = pos)
```
